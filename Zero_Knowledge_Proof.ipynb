```python
import networkx as nx
import random
import matplotlib.pyplot as plt
```

```python
G = nx.Graph() #create graph object
A = list(range(1,31)) # add nodes of group 1( red)
B = list(range(100,131)) # add nodes of group 2 (green)
C = list(range(200,231)) # add nodes of group 3 (blue)
f = plt.figure(figsize=(10,10)) 
for x in range(100): # adding 100 edges, such that there is no edge between 2 nodes from same group. 
    G.add_edge(random.choice(A), random.choice(B)) 
    G.add_edge(random.choice(B), random.choice(A))
    G.add_edge(random.choice(C), random.choice(A))
G.remove_nodes_from(nx.isolates(G)) # removing isolated nodes 
color_map = [] # mapping colors red, blue, and green for groups 1, 2, and 3. 
for node in G:
    if node < 31:
        color_map.append('red')
    elif node < 131: 
        color_map.append('green') 
    elif node < 231:
        color_map.append("blue")
nx.draw(G, node_size = 120, node_color = color_map) # drawing graph G 
plt.show()
    
```
